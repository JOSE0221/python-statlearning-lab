# ¬øQu√© es el Aprendizaje Autom√°tico?

El aprendizaje autom√°tico (Machine Learning, ML), tambi√©n conocido como modelado predictivo en estad√≠stica, 
es un campo que extrae conocimiento de los datos para automatizar la toma de decisiones.  

# Aprendizaje Supervisado

El **aprendizaje supervisado** es el tipo m√°s com√∫n de aprendizaje autom√°tico. Se basa en la construcci√≥n de un **modelo predictivo** a partir de un **conjunto de datos etiquetado**, donde:

- **Entrada (ùë•):** Son las caracter√≠sticas o variables independientes que describen un objeto de inter√©s.  
- **Salida (ùë¶):** Es la variable objetivo o etiqueta que queremos predecir.  

El aprendizaje supervisado consiste en aprender una **funci√≥n matem√°tica** que relacione las entradas con las salidas correctas utilizando un **conjunto de entrenamiento**.

## **1Ô∏è‚É£ Ejemplo Pr√°ctico**
Supongamos que queremos construir un sistema de recomendaciones en una tienda en l√≠nea.  
Cada cliente puede describirse mediante sus caracter√≠sticas:

‚úÖ **Entrada (ùë•)**:  
- Edad  
- Ubicaci√≥n  
- Historial de compras  

‚úÖ **Salida (ùë¶)**:  
- **¬øComprar√° el producto recomendado?** ‚Üí (S√≠/No) ‚Üí **Clasificaci√≥n**  
- **¬øCu√°nto gastar√° en la tienda?** ‚Üí (Cantidad en d√≥lares) ‚Üí **Regresi√≥n**  

Si recopilamos **m√∫ltiples ejemplos hist√≥ricos** de clientes y sus decisiones de compra, podemos construir un **modelo de aprendizaje supervisado**.

## **2Ô∏è‚É£ ¬øPor qu√© se llama "Supervisado"?**
El t√©rmino **"supervisado"** proviene del hecho de que el conjunto de entrenamiento **incluye las respuestas correctas**.  
Durante el proceso de entrenamiento, el modelo ajusta sus par√°metros **comparando sus predicciones con las etiquetas reales** y aprende a hacer mejores predicciones.

---

# **üìå Clasificaci√≥n y Regresi√≥n**
## **1Ô∏è‚É£ Definici√≥n General**
Existen **varios tipos de tareas de aprendizaje supervisado**, como **ranking o estimaci√≥n de probabilidades**, pero nos enfocaremos en las dos m√°s comunes:

1. **Clasificaci√≥n:** Cuando la variable de salida es **discreta** (categor√≠as).
2. **Regresi√≥n:** Cuando la variable de salida es **continua** (valores num√©ricos).

---

## **2Ô∏è‚É£ Clasificaci√≥n**
Un problema de **clasificaci√≥n** ocurre cuando el resultado esperado **es una categor√≠a predefinida**.  

üìå **Ejemplo:**  
- **¬øEs este correo spam?** ‚Üí (S√≠/No)  
- **¬øQu√© tipo de medicamento ser√° m√°s efectivo para un paciente?** ‚Üí (Medicamento A, B o C)  
- **¬øUn cliente comprar√° un producto?** ‚Üí (S√≠/No)  

### **üîπ Clasificaci√≥n Binaria**
Cuando solo hay **dos posibles respuestas**, como **S√≠/No**, se conoce como **clasificaci√≥n binaria**.  
Ejemplos de clasificaci√≥n binaria incluyen:
- **¬øEs un correo spam?** ‚Üí S√≠/No  
- **¬øHay un peat√≥n en la calle?** ‚Üí S√≠/No  
- **¬øUn cliente pagar√° su pr√©stamo?** ‚Üí S√≠/No  

### **üîπ Clasificaci√≥n Multiclase**
Cuando hay **m√°s de dos opciones**, como **tres tipos de medicamentos diferentes**, se llama **clasificaci√≥n multiclase**.  
En este caso, el modelo debe seleccionar la **√∫nica opci√≥n correcta**.

---

## **3Ô∏è‚É£ Regresi√≥n**
Un problema de **regresi√≥n** ocurre cuando el resultado esperado es **una cantidad num√©rica continua**.

üìå **Ejemplo:**  
- **¬øCu√°nto tiempo tardar√° un paciente en recuperarse?** ‚Üí (Medido en d√≠as)  
- **¬øCu√°l ser√° el precio de una casa basada en su ubicaci√≥n y tama√±o?** ‚Üí (Medido en d√≥lares)  
- **¬øCu√°ntas ventas tendr√° un producto el pr√≥ximo mes?** ‚Üí (Medido en unidades vendidas)  

üîπ **Diferencia Clave:**  
Si un paciente debe recibir el alta en **4.5 d√≠as**, pero el modelo predice **5 d√≠as**, el error es peque√±o y la predicci√≥n sigue siendo √∫til.  
En clasificaci√≥n, en cambio, **una predicci√≥n incorrecta es completamente err√≥nea**.

---

## **4Ô∏è‚É£ Diferencias Claves Entre Clasificaci√≥n y Regresi√≥n**
| **Caracter√≠stica**      | **Clasificaci√≥n**                                      | **Regresi√≥n**                                      |
|------------------------|----------------------------------------------------|----------------------------------------------------|
| **Salida (ùë¶)**        | Categor√≠as discretas (ejemplo: "S√≠" o "No")        | Valor continuo (ejemplo: "4.5 d√≠as")              |
| **Ejemplo**           | ¬øEste correo es spam? (S√≠/No)                      | ¬øCu√°ntos d√≠as tardar√° un paciente en recuperarse? |
| **Error aceptable**   | La respuesta debe ser exacta (categor√≠a correcta)  | La predicci√≥n puede estar cerca del valor real    |
| **M√©trica com√∫n**     | Precisi√≥n (% de aciertos)                          | Error cuadr√°tico medio (MSE)                      |

---

## **5Ô∏è‚É£ Casos Ambiguos**
A veces, un problema puede resolverse con **clasificaci√≥n o regresi√≥n**.

üìå **Ejemplo: Calificaciones con Estrellas**  
Supongamos que los usuarios de una aplicaci√≥n pueden calificar un producto con **1 a 5 estrellas**.
- Si lo tratamos como **clasificaci√≥n**, el modelo debe elegir **1, 2, 3, 4 o 5 estrellas**.
- Si lo tratamos como **regresi√≥n**, el modelo puede predecir valores intermedios, como **3.7 estrellas**.

üí° **¬øCu√°l elegir?**  
Depende del contexto y del tipo de modelo que se quiera usar.  
Si el modelo necesita precisi√≥n exacta, la **clasificaci√≥n** es mejor.  
Si el modelo puede hacer predicciones aproximadas, la **regresi√≥n** es m√°s flexible.

# Vecinos Cercanos (K-Nearest Neighbors, K-NN)

El primer algoritmo de clasificaci√≥n que exploraremos es el m√©todo de **Vecinos Cercanos (Nearest Neighbors)**.  
Aunque no se utiliza con frecuencia en la pr√°ctica debido a su **falta de escalabilidad en conjuntos de datos grandes**, es un algoritmo muy intuitivo y, en muchos casos, bastante preciso.

El concepto detr√°s de **K-Nearest Neighbors (K-NN)** es muy simple:
1. **Para clasificar un nuevo punto**, se buscan los **K puntos m√°s cercanos** en el conjunto de datos de entrenamiento.
2. Se asigna la **clase m√°s com√∫n** entre estos **K vecinos** al nuevo punto.

Este m√©todo no requiere una fase de entrenamiento expl√≠cita, ya que simplemente **almacena** los datos de entrenamiento y **clasifica por proximidad** cuando se le da una nueva muestra.

# Control de Aleatoriedad y Estratificaci√≥n en la Divisi√≥n de Datos

Cuando entrenamos un modelo de aprendizaje supervisado, normalmente **dividimos los datos en entrenamiento y prueba**.  
Dado que esta divisi√≥n es **aleatoria**, cada vez que ejecutamos el c√≥digo **los resultados pueden cambiar**.  

Podemos fijar una **semilla aleatoria** para garantizar que los resultados sean **reproducibles**.  
Esto asegura que:
- **Los datos se dividan de la misma manera en cada ejecuci√≥n**.
- **Podamos replicar exactamente el experimento**.

## **1Ô∏è‚É£ ¬øCu√°ndo es Necesario Estratificar?**
La **estratificaci√≥n** es importante cuando nuestra variable objetivo (**target**) tiene una **distribuci√≥n desbalanceada**.  
Por ejemplo:
- Si estamos prediciendo una **enfermedad rara**, y solo el **5% de los pacientes est√°n enfermos**, una divisi√≥n aleatoria podr√≠a generar un conjunto de entrenamiento con **muy pocos casos positivos**, lo que har√≠a que el modelo no aprenda correctamente.
- Si estamos clasificando clientes en **compradores vs. no compradores**, y el **85% de los clientes no compran**, una divisi√≥n aleatoria podr√≠a sesgar la proporci√≥n en los conjuntos de entrenamiento y prueba.

üí° **Soluci√≥n**: Utilizamos la **estratificaci√≥n** para asegurarnos de que la proporci√≥n de clases en el conjunto de entrenamiento sea **igual** a la del conjunto de prueba.

# Divisi√≥n de Datos para Entrenamiento y Prueba

En **aprendizaje autom√°tico**, es fundamental dividir los datos en dos conjuntos:  
1. **Conjunto de entrenamiento (Train Set):** Se usa para **ajustar el modelo** y encontrar los patrones en los datos.  
2. **Conjunto de prueba (Test Set):** Se usa para **evaluar el modelo** con datos que **nunca ha visto** y medir su capacidad de generalizaci√≥n.

## **1Ô∏è‚É£ Controlando el Tama√±o del Conjunto de Prueba**
En Scikit-Learn, podemos controlar el **tama√±o del conjunto de prueba** usando el par√°metro `test_size`.  
Este valor puede ser:
- Un **n√∫mero absoluto de muestras** (ej. `test_size=50` para usar 50 ejemplos como test).  
- Un **porcentaje del conjunto de datos** (ej. `test_size=0.25` para usar el 25% como test).  

üîπ **Regla general:**  
- **75% entrenamiento / 25% prueba** es una recomendaci√≥n com√∫n.  
- **80% entrenamiento / 20% prueba** tambi√©n es ampliamente utilizado.  
- La mejor proporci√≥n **depende del tama√±o del dataset**:  
  - **Si hay pocos datos**, se puede usar m√°s entrenamiento (ej. 90%/10%).  
  - **Si hay muchos datos**, una proporci√≥n 70%/30% podr√≠a ser v√°lida.

## **2Ô∏è‚É£ Haciendo la Divisi√≥n Determin√≠stica con una Semilla Aleatoria**
La divisi√≥n de los datos es **aleatoria**, lo que significa que **cada vez que ejecutemos el c√≥digo, el resultado puede cambiar**.  
Para evitar esto y **hacer nuestros experimentos reproducibles**, fijamos una **semilla aleatoria** (`random_state`).

üí° **¬øPor qu√© fijar una semilla?**
‚úÖ Permite obtener **resultados consistentes** cada vez que ejecutamos el c√≥digo.  
‚úÖ Facilita la **reproducibilidad** del an√°lisis.  
‚úÖ Evita confusi√≥n al comparar diferentes modelos.  

‚ö† **¬øCu√°ndo no fijar una semilla?**
- Si el modelo depende mucho de c√≥mo se dividen los datos, podr√≠a **no ser robusto**.  
- Si solo usamos **una semilla fija**, podr√≠amos haber tenido **suerte o mala suerte** con la divisi√≥n.  
- En an√°lisis m√°s avanzados, es recomendable probar **varias semillas diferentes** y evaluar el promedio de los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Funci√≥n para visualizar la frontera de decisi√≥n
def plot_2d_classification(model, X, ax=None, resolution=100, alpha=0.5, cmap='bwr'):
    """
    Plotea la frontera de decisi√≥n de un clasificador en dos dimensiones.
    
    Par√°metros:
    - model: Clasificador entrenado (Scikit-Learn).
    - X: Datos de entrenamiento (solo con las dos caracter√≠sticas seleccionadas).
    - ax: Objeto de Matplotlib para graficar.
    - resolution: Resoluci√≥n de la malla para el gr√°fico.
    - alpha: Transparencia del color de la frontera de decisi√≥n.
    - cmap: Mapa de colores.
    """
    if ax is None:
        ax = plt.gca()
    
    # Definir los l√≠mites del gr√°fico
    x_min, x_max = X[:, 0].min() - 0.01, X[:, 0].max() + 0.01
    y_min, y_max = X[:, 1].min() - 0.01, X[:, 1].max() + 0.01
    
    # Crear una malla de puntos en el espacio bidimensional
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, resolution),
                         np.linspace(y_min, y_max, resolution))
    
    # Realizar predicciones para cada punto de la malla
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Dibujar la frontera de decisi√≥n usando contourf
    ax.contourf(xx, yy, Z, alpha=alpha, cmap=cmap)
```

```python
# Crear gr√°ficos para visualizar la frontera de decisi√≥n con diferentes valores de K
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

# Probar diferentes valores de K
for ax, n_neighbors in zip(axes.ravel(), [3, 5, 11, 33]):
    ax.set_title(f"n_neighbors={n_neighbors}")
    
    # Entrenar el modelo con el n√∫mero de vecinos seleccionado
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X_train[['mean compactness', 'worst concave points']], y_train)
    
    # Graficar los puntos de entrenamiento
    ax.scatter(X_train['mean compactness'], X_train['worst concave points'], c=y_train, cmap='bwr', s=2)
    
    # Dibujar la frontera de decisi√≥n
    plot_2d_classification(clf, np.array(X_train[['mean compactness', 'worst concave points']]), ax=ax, alpha=.4, cmap='bwr')
    
    # Ajustar la escala de los ejes
    ax.set_aspect("equal")
    ax.set_xlim(0.05, 0.17)
    ax.set_ylim(0.06, 0.2)

plt.tight_layout()
plt.show()
```

## **üìå ¬øC√≥mo Afecta el N√∫mero de Vecinos (K) al Modelo?**

Al observar los gr√°ficos, podemos notar lo siguiente:

üîπ **Cuando K es peque√±o (ej. K=3):**  
- La frontera de decisi√≥n tiene muchas **peque√±as "islas"** de cada clase.  
- El modelo se ajusta mucho a los datos de entrenamiento, capturando detalles locales.  
- Esto puede hacer que el modelo **sea muy preciso en los datos de entrenamiento, pero no generalice bien** en datos nuevos.  

üîπ **Cuando K es grande (ej. K=33):**  
- La frontera de decisi√≥n es **mucho m√°s suave** y ya no hay tantas zonas irregulares.  
- El modelo deja de enfocarse en casos espec√≠ficos y empieza a capturar un **patr√≥n global** en los datos.  
- Sin embargo, si K es demasiado grande, la frontera se vuelve **demasiado simple**, lo que podr√≠a reducir la precisi√≥n.

üîπ **Comportamiento T√≠pico de los Hiperpar√°metros en Machine Learning**  
El n√∫mero de vecinos `K` en **K-Nearest Neighbors** es un **hiperpar√°metro clave**,  
ya que **controla la flexibilidad o complejidad del modelo**:

‚úÖ **K peque√±o (1 o 3):**  
- El modelo es **altamente flexible** y sigue muy de cerca los datos de entrenamiento.  
- Puede capturar patrones detallados, pero tambi√©n **puede sobreajustarse (overfitting)**.  

‚úÖ **K grande (11 o 33):**  
- El modelo es **m√°s general** y sigue un patr√≥n m√°s global.  
- Pierde detalles finos, pero **generaliza mejor a datos nuevos**.  

üí° **Conclusi√≥n:**
- Si **K es muy peque√±o**, el modelo es **demasiado espec√≠fico** y puede fallar con nuevos datos.  
- Si **K es muy grande**, el modelo es **demasiado simple** y puede perder precisi√≥n.  
- Elegir el valor √≥ptimo de `K` es un **balance entre precisi√≥n y generalizaci√≥n**.

## **1Ô∏è‚É£ La Importancia del Escalado**
Las caracter√≠sticas en un conjunto de datos pueden tener **magnitudes muy diferentes**,  
lo que puede afectar el rendimiento de los algoritmos de Machine Learning, especialmente los **basados en distancias** como **K-Nearest Neighbors (K-NN)**.

üí° **Ejemplo:**  
En el conjunto de datos de **c√°ncer de mama**, algunas caracter√≠sticas tienen valores en **cientos o miles**,  
mientras que otras tienen valores entre **0 y 1**.  
Si aplicamos un algoritmo de clasificaci√≥n sin escalar los datos,  
las caracter√≠sticas con valores grandes tendr√°n **m√°s influencia en el modelo**  
solo porque sus valores son mayores en magnitud.

Podemos visualizar esto observando la **distribuci√≥n de las caracter√≠sticas**.

---

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el dataset de c√°ncer de mama
cancer = load_breast_cancer(as_frame=True)
cancer_df = cancer.frame

# Eliminar la variable objetivo para visualizar solo las caracter√≠sticas
cancer_features = cancer_df.drop(columns='target')

# Crear un boxplot para ver las diferencias de escala entre caracter√≠sticas
cancer_features.boxplot(vert=False, figsize=(10, 8))
plt.title("Distribuci√≥n de caracter√≠sticas antes del escalado")
plt.show()
```

## **2Ô∏è‚É£ Diferentes M√©todos de Escalado**

Para corregir este problema, debemos **escalar los datos** para que todas las caracter√≠sticas  
tengan una **escala comparable**.  

Existen varios m√©todos de escalado en Scikit-Learn:

1. **StandardScaler (Estandarizaci√≥n, Z-Score)** ‚Üí Transforma los datos para que tengan **media = 0 y desviaci√≥n est√°ndar = 1**.
2. **MinMaxScaler (Normalizaci√≥n Min-Max)** ‚Üí Escala los datos a un rango espec√≠fico (ej. `[0,1]` o `[-1,1]`).
3. **RobustScaler (Escalado Robusto)** ‚Üí Similar a StandardScaler, pero usa **mediana e IQR** para ser menos sensible a valores at√≠picos.
4. **Normalizer (Normalizaci√≥n de Filas)** ‚Üí Convierte cada fila en un **vector unitario**, √∫til en datos de conteo.

---

```python
from sklearn.preprocessing import StandardScaler

# Crear un escalador est√°ndar
scaler = StandardScaler()

# Ajustar el escalador a los datos de entrenamiento
scaler.fit(X_train)

# Aplicar la transformaci√≥n
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convertir a DataFrame para visualizar
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)

# Ver distribuci√≥n despu√©s del escalado
X_train_scaled_df.boxplot(vert=False, figsize=(10, 8))
plt.title("Distribuci√≥n de caracter√≠sticas despu√©s del escalado")
plt.show()
```

## **3Ô∏è‚É£ C√≥mo Escalar Datos de Prueba de Forma Correcta**
Una regla fundamental en Machine Learning es **NO escalar los datos de prueba de manera independiente**.  

üí° **Ejemplo Incorrecto:**  
Si aplicamos `fit_transform()` a los datos de prueba directamente,  
el modelo podr√≠a aprender **una escala diferente** para `X_test`,  
lo que har√≠a que los valores no sean comparables con `X_train`.

‚úÖ **Soluci√≥n:**  
- **Usar `fit()` solo con `X_train`**.
- **Aplicar `transform()` en `X_test` con los par√°metros aprendidos en `X_train`**.

üìå **Ejemplo Correcto en C√≥digo:**

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Simulaci√≥n de datos de entrenamiento y prueba
X_train = pd.DataFrame({'feature_1': [100, 200, 300, 400, 500],
                        'feature_2': [1, 2, 3, 4, 5]})

X_test = pd.DataFrame({'feature_1': [600, 700, 800],
                       'feature_2': [6, 7, 8]})

# ‚úÖ 1. Crear el escalador
scaler = StandardScaler()

# ‚úÖ 2. Ajustar el escalador SOLO con X_train
scaler.fit(X_train)  

# ‚úÖ 3. Aplicar la transformaci√≥n a X_train y X_test usando los mismos par√°metros
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)  # üö® Usamos transform(), NO fit_transform()

# Convertir a DataFrame para visualizar mejor
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)

# Mostramos los datos escalados
print("X_train escalado:\n", X_train_scaled_df)
print("\nX_test escalado:\n", X_test_scaled_df)
```

# **√Årboles de Decisi√≥n en Machine Learning**

Los **√°rboles de decisi√≥n** son modelos **potentes y flexibles** utilizados tanto en **clasificaci√≥n como en regresi√≥n**.  
Son populares en la industria y ampliamente usados en algoritmos m√°s avanzados como **Random Forests y Gradient Boosting**.

## **1Ô∏è‚É£ ¬øPor qu√© usar √Årboles de Decisi√≥n?**
‚úÖ **Pueden modelar relaciones no lineales** en los datos.  
‚úÖ **No requieren escalar las variables** (a diferencia de K-NN o SVM).  
‚úÖ **Son f√°ciles de interpretar y visualizar**.  
‚úÖ **Pueden manejar datos con valores faltantes** y variables categ√≥ricas.  
‚úÖ **Sirven como base para algoritmos m√°s avanzados (Random Forests, XGBoost, etc.)**.  

```python
# Importar librer√≠as necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# Cargar el dataset de c√°ncer de mama
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)

# Agregar la variable objetivo
df["target"] = cancer.target

# Visualizar primeras filas
print(df.head())

# Gr√°fico de distribuci√≥n de caracter√≠sticas
df.drop(columns="target").boxplot(vert=False, figsize=(10, 8))
plt.title("Distribuci√≥n de Caracter√≠sticas en el Dataset de C√°ncer de Mama")
plt.show()
```

# **Construcci√≥n de √Årboles de Decisi√≥n para Clasificaci√≥n**

Los √°rboles de decisi√≥n clasifican datos **haciendo una serie de preguntas binarias**  
(¬øla caracter√≠stica X es mayor que un umbral?).  
Cada pregunta divide los datos en subconjuntos **cada vez m√°s homog√©neos**,  
hasta que los nodos contienen datos **de una sola clase**.

---

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Dividir el dataset en entrenamiento y prueba (75% - 25%)
X = df.drop(columns="target")
y = df["target"]
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0)

# Crear y entrenar el √°rbol de decisi√≥n
tree = DecisionTreeClassifier(max_depth=3, random_state=0)
tree.fit(X_train, y_train)

# Visualizar el √°rbol
plt.figure(figsize=(12, 6))
plot_tree(tree, feature_names=cancer.feature_names, class_names=["Benigno", "Maligno"], filled=True)
plt.show()
```

## **3Ô∏è‚É£ Construcci√≥n del √Årbol**
El algoritmo construye el √°rbol **de manera recursiva** siguiendo estos pasos:

1Ô∏è‚É£ **Busca la mejor caracter√≠stica y el mejor umbral** para dividir los datos.  
2Ô∏è‚É£ **Minimiza la impureza** de los subconjuntos resultantes.  
3Ô∏è‚É£ **Repite este proceso recursivamente** en cada subconjunto hasta cumplir un criterio de parada.  

---

```python
from sklearn.metrics import accuracy_score

# Predecir en el conjunto de prueba
y_pred = tree.predict(X_test)

# Calcular precisi√≥n
accuracy = accuracy_score(y_test, y_pred)
print(f"Precisi√≥n del √Årbol de Decisi√≥n: {accuracy:.2%}")
```

## **4Ô∏è‚É£ C√≥mo el √Årbol Decide las Preguntas**
Al construir el √°rbol, el modelo intenta minimizar la **impureza** en cada nodo.  
Existen dos m√©todos comunes para medir impureza en clasificaci√≥n:

### **Gini Index (√çndice de Gini)**
\[
H_{gini}(X_m) = \sum_{k} p_{mk} (1 - p_{mk})
\]
- Mide qu√© tan mezcladas est√°n las clases en un nodo.
- Se minimiza cuando **todos los datos en el nodo son de la misma clase**.

### **Cross-Entropy (Entrop√≠a)**
\[
H_{CE}(X_m) = - \sum_{k} p_{mk} \log(p_{mk})
\]
- Similar a Gini, pero con logaritmos.
- Prefiere divisiones donde **una clase es claramente dominante**.

---

```python
# Comparar √°rbol con Gini vs Entrop√≠a
tree_gini = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=0)
tree_entropy = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=0)

tree_gini.fit(X_train, y_train)
tree_entropy.fit(X_train, y_train)

print(f"Precisi√≥n con Gini: {tree_gini.score(X_test, y_test):.2%}")
print(f"Precisi√≥n con Entrop√≠a: {tree_entropy.score(X_test, y_test):.2%}")
```

## **5Ô∏è‚É£ C√≥mo Evitar √Årboles Demasiado Grandes**
Si un √°rbol crece demasiado, puede **memorizar los datos** en lugar de generalizar.  
Para evitarlo, podemos limitar su tama√±o mediante:

‚úÖ **`max_depth`** ‚Üí N√∫mero m√°ximo de niveles en el √°rbol.  
‚úÖ **`min_samples_split`** ‚Üí N√∫mero m√≠nimo de muestras para dividir un nodo.  
‚úÖ **`min_samples_leaf`** ‚Üí N√∫mero m√≠nimo de muestras en una hoja.  

---

```python
# Limitar la profundidad del √°rbol
tree_pruned = DecisionTreeClassifier(max_depth=4, min_samples_split=10, random_state=0)
tree_pruned.fit(X_train, y_train)

# Evaluaci√≥n del modelo podado
print(f"Precisi√≥n del √Årbol Podado: {tree_pruned.score(X_test, y_test):.2%}")
```
